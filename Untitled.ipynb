```python
hwy = df.hwy_mpg.mean()
```

```python
hwy
```

```python
df["hwy_mpg"] = df.hwy_mpg.fillna(hwy)
```

```python
df.hwy_mpg.isnull().sum()
```

```python
lengthee = df.length.mean()
```

```python
lengthee
```

```python
df["length"] = df.length.fillna(lengthee)
```

```python
df.length.isnull().any()
```

```python
weighttt = df.weight.mean()
```

```python
weighttt
```

```python
df["weight"] = df.length.fillna(weighttt)
```

```python
df.weight.isnull().any()
```

```python
wheel_baseee = df.wheel_base.mean()
```

```python
df["wheel_base"] = df.length.fillna(wheel_baseee)
```
